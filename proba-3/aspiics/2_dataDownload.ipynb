
Before we start we will import the needed packages


````python
import requests
````

Generate the API request

````python
# baseAPI url
baseAPI = "https://p3sc.oma.be/api/"
# Level of the data
level = "L2"
# select the columns
select = "select=name,version"
# add the fixed filter
fixed_selector = "active=is.true&version=eq.v2"
# add variable filter options
variable_filter = "FILTER=like.Wide*"
# limit the amount of results
limit = "limit=10"

#generate the final url
apiQuery = f"{baseAPI}{level}?{select}&{fixed_selector}&{variable_filter}&{limit}"

apiQuery
````

Connect with the API using requests

````python
response = requests.get(apiQuery)
data = response.json()
data
````

Download the data based on the result of the API.

````python
fileURLList = []
baseDataURL = "https://p3sc.oma.be/datarepfiles"
fileLevel = "L2"
for item in data:
    file_name = item['name']
    file_version = item['version']
    file_url = f"{baseDataURL}/{fileLevel}/{file_version}/{file_name}"
    fileURLList.append((file_name, file_url))
fileURLList
````

Download the data

````python
import os
output_folder = "output"
os.makedirs(output_folder, exist_ok=True)
    
for file_name, file_url in fileURLList:
    print(f"Downloading {file_name} from {file_url}")
    file_response = requests.get(file_url)
    with open(os.path.join(output_folder, file_name), 'wb') as f:
        f.write(file_response.content)
    print(f"Downloaded {file_name}")
````
