```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import aspiics_colormaps

filename = "aspiics_l3_bt_1590400D0006_20250913T140854.fits"

with fits.open(filename) as hdul:
    data = hdul[0].data

data = np.array(data, dtype=float)

plt.figure(figsize=(6, 6))
plt.imshow(np.sqrt(data), origin="lower", cmap="bt")
plt.colorbar(label="sqrt(MSB)")
plt.title(filename)
plt.tight_layout()
plt.show()
```
